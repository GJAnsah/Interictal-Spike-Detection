```python
import pandas as pd
from scipy.signal import *
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import decimal
import mne
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = 10, 5
plt.rcParams['font.size'] = 12
```

```python
data = pd.read_csv('processed_data.csv')

times=data['time']
Ptimes=times.to_numpy()

data=data.drop('time', axis=1)
Pdata=data.to_numpy()
Pdata=Pdata.transpose()

sr=500
```

```python
for i in range(len(Pdata)):
    x=Pdata[i][110*sr:116*sr]
    norm = np.linalg.norm(x)
    x = x/norm
    t=Ptimes[110*sr:116*sr]
    peaks, _ = find_peaks(x, rel_height=0.5,prominence=1.5*np.mean(x),height=[0.015,0.08],width=[0.02*sr,0.2*sr],distance=sr*0.2)
    
    plt.subplot(20,1, i+1)
    plt.plot(t,x)
    plt.plot(t[peaks], x[peaks], "o")
#     plt.axis('off')
    plt.show()
```

```python
exp_var_pca=[0.29836956, 0.22173101, 0.07725535, 0.07370767, 0.06166612,
       0.03936943, 0.03280752, 0.02987992, 0.02480109, 0.02083844,
       0.02005489, 0.01529161, 0.01282643, 0.01120994, 0.00838531,
       0.00772833, 0.00623471, 0.00558897, 0.00497124, 0.00450004]

cum_sum_eigenvalues = np.cumsum(exp_var_pca)
fig, ax = plt.subplots(figsize = (10,4))
plt.bar(range(1,len(exp_var_pca)+1), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1,len(cum_sum_eigenvalues)+1), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
idx=np.asarray(list(range(1,21)))
ax.set_xticks(idx)
plt.legend(loc='center')
plt.tight_layout()
plt.show()
```
